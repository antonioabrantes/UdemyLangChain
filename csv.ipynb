```python
texto="""básico,janeiro,200
médio,fevereiro,300
médio,março,400"""

data=[]
for line in texto.splitlines():
    values = line.split(',')
    print(values)
    data.append(values)

print(data)
```

```python
import pandas as pd
df = pd.DataFrame(data,columns=['Modo','Mês','Valor'])
display(df)
```

```python
# https://www.kaggle.com/datasets/sahirmaharajj/library-services-dataset
import pandas as pd
file_path = 'bib.csv'
data=[]
with open(file_path,'r',encoding='utf-8') as file:
    for line in file:
        values = line.strip().split(',')
        data.append(values)

headers = data[0]
rows = data[1:]
df = pd.DataFrame(rows,columns=headers)
display(df)
```
