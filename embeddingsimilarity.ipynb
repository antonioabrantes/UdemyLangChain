```python
# as dimensoes dos dois vertores  diferente, isso causa erro em similaridade

import numpy as np

# Supondo que embed1 e embed2 sejam listas de vetores com tamanhos diferentes
embed1 = embeddings.embed_documents("coca-cola")
embed2 = embeddings.embed_documents("cappuccino")
embed1 = np.mean(embed1, axis=0)
embed2 = np.mean(embed2, axis=0)

# Agora, as dimenses devem ser iguais, e a similaridade pode ser calculada
similarity = cosine_similarity([embed1], [embed2])
print(similarity[0][0])  # Similaridade como um nmero
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
embed1 = embeddings.embed_documents("coca-cola")
embed2 = embeddings.embed_documents("telhado")
# Supondo que embed1 e embed2 sejam listas
embed1 = np.array(np.mean(embed1, axis=0))  # Converte para um array NumPy
embed2 = np.array(np.mean(embed2, axis=0))  # Converte para um array NumPy

# Calcula a similaridade do cosseno. reshape existente em um np.array Ajusta a forma do array para que 
# seja uma matriz 2D com uma linha e o nmero apropriado de colunas, necessrio para a funo cosine_similarity
similarity = cosine_similarity(embed1.reshape(1, -1), embed2.reshape(1, -1))

print("Similaridade:", similarity[0][0])
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Initialize the model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Embed the texts
embedding1 = model.encode("coca-cola")
embedding2 = model.encode("pepsi")

# Calculate cosine similarity
similarity = cosine_similarity(
    embedding1.reshape(1, -1),
    embedding2.reshape(1, -1)
)[0][0]

print("Cosine similarity:", similarity)
print("Length of coca embedding:", len(embedding1))
print("Length of cappuccino embedding:", len(embedding2))
```
